```python
class calculator:
    def __init__(self,num1,num2):
        self.num1=num1
        self.num2=num2
  
    def add (self):
        print(f" sum of {self.num1} and {self.num2} is {self.num1+self.num2}")
    def sub (self):
        print(f" sub of {self.num1} and {self.num2} is {self.num1-self.num2}")
    def mul (self):
        print(f" mul of {self.num1} and {self.num2} is {self.num1*self.num2}")
    def div (self):
        print(f" div of {self.num1} and {self.num2} is {self.num1/self.num2}")

num=calculator(10,20)
num.add()
num.sub()
num.mul()
num.div()
```

```python
class temp:
    def __init__ (self,C):
        self.C=C
    def CtoK(self):
        print(f"temp in K: {self.C+273}")
    def CtoF(self):
        print(f"temp in F: {(self.C*(9/5))+32}")
temp1=temp(37)
temp1.CtoK()
temp1.CtoF()
    
```

```python
class library:
    def __init__(self):
        self.books=["1984","The 48 Laws Of Power"]
    def addbook (self,book):
        if book not in self.books:
            self.books.append(book)
        else:
            print("The book is already available in the library")
    def borrow (self,book):
        self.books.remove(book)

    def returnbook (self,book):
        self.books.append(book)
        print("The book is returned in the library")

    def showbooks (self):
        print(f"The available books are: {self.books}")
        
library1=library()
library1.showbooks()
```

```python
class votingsys:
    def __init__(self):
        self.candidates=[]
        self.votes=[]
    def vote(self,candidate):
        if candidate in self.candidates:
            index= self.candidates.index(candidate)
            self.votes[index]+=1
        else:
            self.candidates.append(candidate)
            self.votes.append(1)
    def result (self):
        for i in range(len(self.candidates)):
            print(f"{self.candidates[i]}: {self.votes[i]}")

v1=votingsys()
v1.vote("Trump")
v1.vote("Trump")
v1.vote("Bayden")
v1.vote("Trump")
v1.vote("Bayden")
v1.result()
        
```
